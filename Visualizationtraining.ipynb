# **Training Visualization**

# Bar diagram

```python
#bar diagram with maplotlib (PYPLOT) and numpy
#bar chart training
#possivel atu plota valor zero (0) ba iha barra
#vertical bar

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A","B","C","D","E","F"])
y = np.array([3,8,1,10,0,12])

plt.title("Bar chart")
plt.bar(x,y)
plt.show()
```

```python
#bar diagram with maplotlib (PYPLOT) and numpy
#bar chart training
#possivel atu plota valor zero (0) ba iha barra
#horizontal  bar

import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A","B","C","D","E","F"])
y = np.array([3,8,1,10,1,12])

plt.title("Bar chart")
plt.barh(x,y)
plt.show()
```

```python
#vertical bar color name
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 2, 10])

plt.title("Tau cor iha barra")
plt.bar(x, y, color = "red")
plt.show()
```
